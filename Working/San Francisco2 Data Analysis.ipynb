```python
%matplotlib inline
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')

sf_clean=pd.read_csv("../Resources/San Francisco/sanfrancisco_clean.csv")
sf_clean
```

```python
#get the price vs overall satisfaction
sf_price=sf_clean["price"]
sf_sat=sf_clean["overall_satisfaction"]


plt.scatter(sf_price,sf_sat,marker="o",edgecolor="black",)
sns.set(rc={'figure.figsize':(7,5)})
plt.title("San Francisco Price vs Satisfaction")
plt.xlabel("Price")
plt.ylabel("Satisfaction")
plt.savefig("sf_figures/sf_priceVsatis0.png")
```

```python
#get room types  vs Average Price
    #find out the room type
sf_rooms=sf_clean["room_type"].unique()

sf_roomlist=[]
sf_roommean=[]
#get the list of unique rooms
for x in sf_rooms:
    sf_roomlist.append(x)
    #get the price mean of the room type 
    sf_roommean.append(sf_clean.loc[sf_clean["room_type"]==x]["price"].mean())
#print(sf_roomlist)
#print(sf_roommean)
#plot them 
sns.set(rc={'figure.figsize':(7,5)})

#add colors
colors = ['#ff9999','#66b3ff','#99ff99',]

plt.bar(sf_roomlist,sf_roommean,align="center", alpha=0.9, color=colors)
plt.title("San Francisco Room Type vs Average Price")
plt.xlabel("Room Type")
plt.ylabel("Average Price")
plt.grid(True)
plt.savefig("sf_figures/sf_roomVprice.png")
```

```python
sf_roomsat=[]
#get satisfaction mean
for x in sf_rooms:
    sf_roomsat.append(sf_clean.loc[sf_clean["room_type"]==x]["overall_satisfaction"].mean())
sf_roomsat
print(sf_roomsat)

#plot them 
sns.set(rc={'figure.figsize':(7,5)})

#add colors
colors = ['#ff9999','#66b3ff','#99ff99',]




plt.bar(sf_roomlist,sf_roomsat,align="center", alpha=0.9, color=colors,)
plt.title("San Francisco Room Type vs Satisfaction")
plt.xlabel("Room Type")
plt.ylabel("Average Satisfaction")
plt.grid(True)
plt.tight_layout()
plt.savefig("sf_figures/sf_roomVsatisfaction.png")
```

```python
sf_hood =sf_clean["neighborhood"].unique()

sf_hoodlist=[]
sf_hoodmean=[]
#get the list of unique rooms
for x in sf_hood:
    sf_hoodlist.append(x)
   #get the price mean of the room type
    sf_hoodmean.append(sf_clean.loc[sf_clean["neighborhood"]==x]["price"].mean())
#print(sf_hoodlist)
#print(sf_hoodmean)
#Set tick colors:
sns.set(rc={'figure.figsize':(8, 20)})
#plot them

#add colors
colors = ['#66b3ff','#99ff99']

plt.barh(sf_hoodlist,sf_hoodmean, align='center', alpha=0.9,color=colors,)
plt.title('San Francisco Neighborhood vs Average Price')
plt.xlabel('Average Price')
plt.ylabel('Neighborhood')
plt.grid(True)
plt.savefig("sf_figures/sf_neighborhoodVprice.png")
```

```python
#overall Price average for boston
SF_Price=sf_clean["price"].mean()
print(f"San Francisco total average price is : {SF_Price}")
```

```python
#count the room type
sf_shared=sf_clean.loc[sf_clean["room_type"]=="Shared room"]["room_type"].count()
sf_entire=sf_clean.loc[sf_clean["room_type"]=="Entire home/apt"]["room_type"].count()
sf_private=sf_clean.loc[sf_clean["room_type"]=="Private room"]["room_type"].count()


sf_clean["room_type"].unique()
sf_pie_rooms=[sf_shared,sf_entire,sf_private]
sf_pie_label=["Shared room","Entire home/apt","Private room"]

#add colors
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']


print("Number of rooms in San Francisco")
print(f"Shared rooms: {sf_shared}")
print(f"Entire home/spt rooms: {sf_entire}")
print(f"Private rooms: {sf_private}")

explode=(0.1,0.1,0.1)
plt.pie(sf_pie_rooms,explode=explode,labels=sf_pie_label,colors=colors,shadow=True,
       autopct="%1.1F%%",startangle=90,) 

#draw circle
centre_circle = plt.Circle((0,0),0.50,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal') 
plt.tight_layout()
plt.savefig("sf_figures/sf_rooms.png")
```

```python
#set the x axis
sfpricelist=["0~100","101~200","201~300","301~400","401~500","501~600","601~700"
           ,"701~800","801~900","901~1000","1000up"]
#sf satisfaction average sfy range
sfsat0=sf_clean.loc[(sf_clean["price"]<101)&(sf_clean["price"]>0)]["overall_satisfaction"].mean()
sfsat100=sf_clean.loc[(sf_clean["price"]>100)&(sf_clean["price"]<201)]["overall_satisfaction"].mean()
sfsat200=sf_clean.loc[(sf_clean["price"]>200)&(sf_clean["price"]<301)]["overall_satisfaction"].mean()
sfsat300=sf_clean.loc[(sf_clean["price"]>300)&(sf_clean["price"]<401)]["overall_satisfaction"].mean()
sfsat400=sf_clean.loc[(sf_clean["price"]>400)&(sf_clean["price"]<501)]["overall_satisfaction"].mean()
sfsat500=sf_clean.loc[(sf_clean["price"]>500)&(sf_clean["price"]<601)]["overall_satisfaction"].mean()
sfsat600=sf_clean.loc[(sf_clean["price"]>600)&(sf_clean["price"]<701)]["overall_satisfaction"].mean()
sfsat700=sf_clean.loc[(sf_clean["price"]>700)&(sf_clean["price"]<801)]["overall_satisfaction"].mean()
sfsat800=sf_clean.loc[(sf_clean["price"]>800)&(sf_clean["price"]<901)]["overall_satisfaction"].mean()
sfsat900=sf_clean.loc[(sf_clean["price"]>900)&(sf_clean["price"]<1001)]["overall_satisfaction"].mean()
sfsat1000up=sf_clean.loc[(sf_clean["price"]>1000)]["overall_satisfaction"].mean()
sfsatlist=[sfsat0,sfsat100,sfsat200,sfsat300,sfsat400,sfsat500,sfsat600,sfsat700
            ,sfsat800,sfsat900,sfsat1000up]



#plot them 
sns.set(rc={'figure.figsize':(7,5)})

#add colors
colors = ['#66b3ff','#99ff99',]







plt.bar(sfpricelist,sfsatlist,align="center", alpha=0.9, color=colors,)
plt.xticks(rotation=45)
plt.title("San Francisco Price vs Average Satisfaction")
plt.xlabel("Price Range")
plt.ylabel("Average Satisfaction")
plt.grid(True)
plt.tight_layout()
plt.savefig("sf_figures/sf_priceVsatis_mean.png")
```
