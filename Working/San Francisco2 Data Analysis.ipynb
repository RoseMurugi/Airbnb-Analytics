```python
%matplotlib inline
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')

sf_clean=pd.read_csv("../Resources/san_francisco/san_francisco_cleandata.csv")
sf_clean
```

```python
#get the price vs overall satisfaction
sf_price=sf_clean["price"]
sf_sat=sf_clean["overall_satisfaction"]
plt.scatter(sf_price,sf_sat,marker="o",edgecolor="black",)
sns.set(rc={'figure.figsize':(8,5)})
plt.title("San Francisco Price vs Satisfaction")
plt.xlabel("Price")
plt.ylabel("Satisfaction")
plt.show()

#plt.savefig("sf_figures/sf_priceVsatis0.png")
```

```python
#get room types  vs price and satisfaction
    #find out the room type
sf_rooms=sf_clean["room_type"].unique()

sf_roomlist=[]
sf_roommean=[]
#get the list of unique rooms
for x in sf_rooms:
    sf_roomlist.append(x)
    #get the price mean of the room type 
    sf_roommean.append(sf_clean.loc[sf_clean["room_type"]==x]["price"].mean())
#print(sf_roomlist)
#print(sf_roommean)
#plot them 
sns.set(rc={'figure.figsize':(8,5)})
plt.bar(sf_roomlist,sf_roommean,align="center", alpha=0.9, color=["blue"])
plt.title("San Francisco Room Type vs Average Price")
plt.xlabel("Room Type")
plt.ylabel("Average Price")
plt.show()
#plt.savefig("sf_figures/sf_roomVprice.png")
```

```python
sf_hood =sf_clean["neighborhood"].unique()

sf_hoodlist=[]
sf_hoodmean=[]
#get the list of unique rooms
for x in sf_hood:
    sf_hoodlist.append(x)
   #get the price mean of the room type
    sf_hoodmean.append(sf_clean.loc[sf_clean["neighborhood"]==x]["price"].mean())
#print(sf_hoodlist)
#print(sf_hoodmean)
#Set tick colors:
sns.set(rc={'figure.figsize':(8, 20)})
#plot them
plt.barh(sf_hoodlist,sf_hoodmean, align='center', alpha=0.9,color=["green", "blue"])
plt.title('San Francisco Neighborhood vs Average Price')
plt.xlabel('Average Price')
plt.ylabel('Neighborhood')
plt.show()
```
