```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
from IPython.display import Image
```

```python
#import csvs of data to merge
AMS=pd.read_csv("../Resources/Amsterdam/amsterdam_clean.csv")
#AMS.head()
Boston=pd.read_csv("../Resources/boston/boston_clean.csv")
#Boston.head()
Nairobi=pd.read_csv("../Resources/Nairobi/nairobi_clean.csv")
#Nairobi.head()
SF=pd.read_csv("../Resources/San Francisco/sanfrancisco_clean.csv")
#SF.head()
```

```python
#merging the four data frames
combined_city_data = pd.concat([AMS, SF,Boston,Nairobi], axis=1)
combined_city_data.head()
```

```python
# Set the 'city' to be our index for easy referencing of rows
Cty=combined_city_data.dropna(axis="columns", how="any")
Cty=Cty.dropna(axis="rows", how="any")
Cty=Cty.drop(columns=["latitude","longitude","bedrooms"])
Cty
```
