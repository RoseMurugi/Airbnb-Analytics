```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


sf_data = "../Resources/san_francisco/san_francisco.csv"

sf_data_df = pd.read_csv(sf_data)

sf_data_df.head()
```

```python
#clean the empty column and rows:
sf_data_df=sf_data_df.dropna(axis="columns", how="any")
sf_data_df=sf_data_df.dropna(axis="rows", how="any")
sf_data_df=sf_data_df.drop(columns=["room_id","survey_id","host_id","last_modified"])
sf_data_df.head()
```

```python
#To save the cleaned data:
sf_data_df.to_csv("../Resources/san_francisco/san_francisco_cleandata.csv")
```
