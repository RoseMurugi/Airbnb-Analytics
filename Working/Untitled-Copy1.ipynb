```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
#first read the csv:
Nrbdata=pd.read_csv("../Resources/Nairobi/tomslee_airbnb_nairobi_1530_2017-07-25.csv")
#clean the empty column and rows:
Nrb=Nrbdata.dropna(axis="columns", how="any")
Nrb=Nrb.dropna(axis="rows", how="any")
Nrb=Nrb.drop(columns=["room_id","survey_id","host_id","last_modified"])
```

```python
Nrb
```

```python
#saving cleaned csv
Nrb.to_csv("Nairobi_cleandata.csv")
```

```python
#calculating the mean price
nrb_mean=Nrb["price"].mean()
nrb_mean
```

```python
nrb_roomlist=[]
nrb_roommean=[]
nrb_rooms=Nrb["room_type"].unique()
for x in nrb_rooms:
    nrb_roomlist.append(x)
    nrb_roommean.append(Nrb.loc[Nrb["room_type"]==x]["price"].mean())
print(nrb_roomlist)
print(nrb_roommean)
```

```python
Rooms= nrb_roomlist
x_axis=np.arange(len(Rooms))

Mean = nrb_roommean
#x_axis = np.arange(len(Rooms))

plt.bar(x_axis,Mean,color='g',align='center',alpha=0.5)
plt.xticks(x_axis,Rooms)
plt.ylabel("Mean Price")
plt.title("Price Vs Room Type")

plt.show()

```
